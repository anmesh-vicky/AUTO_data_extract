```python
def normdata(data):
    
    (len1,w1) = np.shape(data)
    ndata = np.zeros([len1,w1//2])
    for i in range(w1//2):
        ndata[:,i]=(data[:,2*i+1]-min(data[:,2*i+1]))/(max(data[:,2*i+1])-min(data[:,2*i+1]))
    return ndata
```

```python
enc = OneHotEncoder(sparse=False)
train_y=enc.fit_transform(df_2.reshape(-1,1))
train_y.shape
```

```python
X_train ,X_test ,y_train,y_test = train_test_split(train,train_y,test_size = 0.2,shuffle=True ,random_state = 42)
```

```python
# Neural networks uses Keras with TF background
import keras as K
from keras.models import Model
from keras.models import Sequential
from keras.layers import Dropout
from keras.callbacks import ReduceLROnPlateau
from keras.callbacks import EarlyStopping
from keras.layers import  MaxPooling1D
import tensorflow as tf
from keras.callbacks import TensorBoard
from sklearn import metrics
from keras.regularizers import l1
# Clear Keras and TF session, if run previously
K.backend.clear_session()
tf.reset_default_graph()
config = tf.ConfigProto()
config.gpu_options.allow_growth = True
sess = tf.Session(config=config)
# Training Parameters

accuracy_test=[]
accuracy_train=[]
logs=[]
predictions_ord=[]
matrix=[]
# Network Parameters
n_input = 400 # Total angles in XRD pattern
n_classes = 3 # Number of space-group classes
# Define network structure
model = Sequential()

model.add(K.layers.Conv1D(64, 8,strides=3, padding='same',input_shape=(400,1), activation='relu'))
#model.add(Dropout(0.2))
#model.add(K.layers.MaxPooling1D(pool_size=10, stride=1))
model.add(K.layers.Conv1D(64, 5,strides=2, padding='same', activation='relu'))
#model.add(K.layers.MaxPooling1D(pool_size=10, stride=1))
#model.add(Dropout(0.2))
model.add(K.layers.Conv1D(64, 3,strides=1, padding='same', activation='relu'))
#model.add(K.layers.MaxPooling1D(pool_size=10, stride=1))
#model.add(Dropout(0.2))
#model.add(K.layers.Conv1D(64, 3,strides=1, padding='same', activation='relu'))
#model.add(K.layers.MaxPooling1D(pool_size=10, stride=1))
#model.add(K.layers.Conv1D(64, 3,strides=1, padding='same', activation='relu'))
#model.add(Dropout(0.2))
#model.add(K.layers.MaxPooling1D(pool_size=10, stride=1))
#model.add(K.layers.Conv1D(32, 3,strides=1, padding='same', activation='relu'))
# model.add(K.layers.MaxPooling1D(pool_size=2, stride=1))
#model.add(K.layers.Conv1D(32, 3,strides=1, padding='same', activation='relu'))
#model.add(K.layers.Conv1D(32, 3,strides=1, padding='same', activation='relu'))
#model.add(K.layers.Conv1D(32, 3,strides=1, padding='same', activation='relu'))
#model.add(K.layers.MaxPooling1D(pool_size=5, stride=1))
model.add(K.layers.GlobalAveragePooling1D())
#model.add(K.layers.LSTM(32,input_shape=X_train.shape, return_sequences=False))
model.add(K.layers.Dense(n_classes, activation='softmax'))
        
#Define optimizer        
optimizer = K.optimizers.Adam()
#tensorboard = TensorBoard(log_dir='./logs', histogram_freq=0,
#                          write_graph=True, write_images=False)   
# Compile model
model.compile(loss='categorical_crossentropy',
                      optimizer=optimizer,
                      metrics=['categorical_accuracy'])
# Fit model
hist = model.fit(X_train, y_train, batch_size=32, nb_epoch=100,verbose=1, validation_data=(X_test, y_test))
prediction=model.predict(X_test)
 
#Go from one-hot to ordinal...
prediction_ord=[np.argmax(element) for element in prediction]
predictions_ord.append(prediction_ord)
matrix = metrics.confusion_matrix(y_test.argmax(axis=1), prediction.argmax(axis=1))
#Compute loss and accuracy for each k validation
accuracy_test=model.evaluate(X_test, y_test, verbose=1)
accuracy_train=model.evaluate(X_train, y_train, verbose=1)
        
#Save logs
log = pd.DataFrame(hist.history)
logs.append(log)
print(accuracy_train[1], accuracy_test[1]) 
```

```python
plt.plot(logs[0]['categorical_accuracy'], 'C2', label = 'train')
plt.plot(logs[0]['val_categorical_accuracy'], 'C3', label = 'test')
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend( loc='upper left')
plt.show()
```
